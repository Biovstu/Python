```python
import sympy 
```

```python
x = sympy.Symbol('x')
```

```python
f_x = 32 * x * x* x - 12 * x**2 + 35
print(f_x)
```

```python
print(f_x.subs(x, 5))
```

```python
sympy.plotting.plot(f_x, (x, -1, 1))
```

```python
diff_f_x = f_x.diff(x)
```

```python
sympy.plotting.plot(diff_f_x, (x, -1, 1))
```

```python
diff_zeros = sympy.solve(diff_f_x, x)
```

```python
for zero in diff_zeros:
    print(f'f({zero}) = {f_x.subs(x, zero)}')
```
